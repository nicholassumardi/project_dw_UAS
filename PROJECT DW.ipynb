# Grafik Rata-Rata Nilai Berdasarkan Gender

```python
plt.figure(figsize=(10,6))
plt.title("Rata-Rata Nilai Berdasarkan Gender")
df_melted = df.melt(id_vars=['gender'], value_vars=['math score', 'writing score', 'reading score'], var_name='score name', value_name='nilai rata-rata')

sns.barplot(data=df_melted, x='gender', y='nilai rata-rata', hue='score name')
```

# Grafik Rata-Rata Nilai Berdasarkan Ras

```python
plt.figure(figsize=(10,6))
plt.title("Rata-Rata Nilai Berdasarkan Ras")
df_melted2 = df.melt(id_vars=['race/ethnicity'], value_vars=['math score', 'writing score', 'reading score'], var_name='score name', value_name='nilai rata-rata')

sns.barplot(data=df_melted2, x='race/ethnicity', y='nilai rata-rata', hue='score name')
```
