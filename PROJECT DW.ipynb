# RATA-RATA NILAI BERDASARKAN RAS

```python
print('Nilai Mat \n')
print(df.groupby(['race/ethnicity'])['math score'].mean())
print('\n')
print('Nilai Baca \n')
print(df.groupby(['race/ethnicity'])['reading score'].mean())
print('\n')

print('Nilai Tulis \n')
print(df.groupby(['race/ethnicity'])['writing score'].mean())
```

# Grafik Rata-Rata Nilai Matematika Berdasarkan Gender

```python
plt.figure(figsize=(10,6))
plt.title("Rata-Rata Nilai Matematika Berdasarkan Gender")
sns.barplot(x=data['gender'], y=data['math score'])
```

# Grafik Rata-Rata Nilai Baca Berdasarkan Gender

# Grafik Rata-Rata Nilai tulis Berdasarkan Gender

```python
plt.figure(figsize=(10,6))
plt.title("Rata-Rata Nilai tulis Berdasarkan Gender")
sns.barplot(x=data['gender'], y=data['writing score'])
```

# Grafik Hubungan antara nilai matematika dengan Tingkat Pendidikan Orangtua

```python
plt.figure(figsize=(10,6))
plt.title("Hubungan antara nilai matematika dengan Tingkat Pendidikan Orangtua")
sns.swarmplot(x=data['parental level of education'],
              y=data['math score'], hue=data['parental level of education'])
plt.xticks(rotation=90)
```

# Grafik Hubungan antara nilai membaca dengan Tingkat Pendidikan Orangtua

```python
plt.figure(figsize=(10,6))
plt.title("Hubungan antara nilai membaca dengan Tingkat Pendidikan Orangtua")
sns.swarmplot(x=data['parental level of education'],
              y=data['reading score'], hue=data['parental level of education'])
plt.xticks(rotation=90)
```

# Grafik Hubungan antara nilai menulis dengan Tingkat Pendidikan Orangtua
